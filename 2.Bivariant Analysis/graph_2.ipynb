```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance.csv",na_values=(['!','@','#','$','%','^','&','*'," ","none"]))
```

```python
min(data['reading score'])
```

```python
sns.scatterplot(data=data ,x='writing score',y='reading score',hue=data['gender'])
```

```python
bar_plot=sns.barplot(x='parental level of education',y='math score',data=data,errwidth=0)
for i in bar_plot.containers:
    bar_plot.bar_label(i,)
```
